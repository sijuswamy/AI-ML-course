# Basic Matrix Operations

## Print outputs in python

Use the `print()` command to display content of a variable in python. For example a message 'Hello Python' can be displayed as:

```python
print("Hello Python")
```

Displaying output of a variable and operation in python as follows:

```python
a=2
b=5
print("Value of a:",a)
print("Value of b: ", b)
print("Value of the operation, a+(2b) is:",a+(2*b))
```

## Reading a vector in `python`

```python
import numpy as np
v1=np.array([1,2,3])
v2=np.array([2,4,5])
a=2.3
print("The vector is: {}".format(v1))
print('The scalar product of {}  with {} is : {}'.format(v1,a,a*v1))
print('The linear combination of {} and {} is :{}'.format(v1,v2,a*v1+v2))
```

## Matrix operations in `python`

```python
M1=np.array([[1,2,3],[2,3,4],[4,5,6]])
M2=np.array([[3,2,1],[2,4,6],[5,6,7]])
print('Sum of matrices:\n',M1+M2)
print('Scalar product:\n',a*M1)
print('Element-wise product:\n',M1*M2)
print('Matrix Product\n',np.dot(M1,M2))
```
